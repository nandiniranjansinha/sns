1. distplot (Deprecated)

- What it is:
  An older function that combined both histogram and KDE plotting into one.

- Key points:
  - Can plot histogram and/or KDE on the same axes.
  - Flexible but limited to one plot on one axes.
  - Deprecated as of Seaborn 0.11 and removed in 0.14.
  - Raises warnings if used now.

- Example:
`sns.distplot(data, bins=20, kde=True)`

2. histplot (Axes-level function)

- What it is: A newer function designed specifically for histograms and optionally KDE.

- Key points:
  - Plots histogram on a single axes.
  - Can overlay KDE by setting kde=True.
  - More features and better integration with Seabornâ€™s new API.
  - Replaces most uses of distplot when you want a histogram.

- Example:
`sns.histplot(data, bins=20, kde=True)`

3. displot (Figure-level function)

- What it is:
A figure-level function that creates a figure with multiple facets (subplots) if desired.

- Key points:
  - Creates a FacetGrid by default, useful for plotting distributions by subsets of data.
  - Can plot histogram, KDE, or both.
  - Supports faceting with col=, row=, hue= parameters.
  - More flexible for exploratory data analysis with multiple groups.

- Example:
`sns.displot(data, bins=20, kde=True)` or with faceting:

  `sns.displot(data, col="category", kde=True)`

## Bar Plot

```python
sns.barplot(x = 'sex', y = 'tip', data = tips, palette = 'pastel')
```

**The warning means:**

- You passed a palette argument to sns.barplot() without using a hue parameter.

- This usage is deprecated and will be removed in Seaborn version 0.14.0.

- Seaborn expects you to use palette only when you have a hue variable to color different groups.

- If you want colors for the bars without grouping by hue, Seaborn suggests a workaround:

      Set hue='sex' (same as x variable), and

      Set legend=False to avoid showing the legend.

**Why this happens**

- Palette is meant to specify colors for groups created by the hue variable.

- Without hue, Seaborn is warning that using palette is not the intended way to color bars.

```python
sns.barplot(x='sex', y='tip', data=tips, hue='sex', palette='pastel', legend=False)
```

```python
sns.barplot(x='sex', y='tip', data=tips, hue='sex', palette='pastel')
```
